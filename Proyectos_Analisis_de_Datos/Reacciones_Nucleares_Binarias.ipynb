# Reaccciones Nucleares Binarias 

El proposito de este notebook es el de encontrar todas las reacciones nucleares binarias tal que produzcan ${^{26}_{13}Al}$

$$A+b \rightarrow c + {^{26}_{13}Al}$$.


````python
import pandas as pd
import numpy as np
````

Importamos una base de datos para encontrar la masa y propiedades de los distintos nucleotidos.
Sacada de https://github.com/ffavela/isonav

````python
datos = pd.read_csv("ATOMIC.MASS.csv", sep=',')
datos
````

````python
DATOS=datos.drop(labels=0)
#Cambiamos el tipo de dato de esta columna para operar con ella (ya que tiene las masas)
DATOS["ATOMIC MASS"]=DATOS["ATOMIC MASS"].astype("int")
DATOS.head()
````

Buscamos las propiedades del ${^{26}_{}}Al$

````python
DATOS[DATOS["A"]==26]
````

Definimos la funciÃ³n para el calculo de energÃ­a.
$$Q=E_f-E_i$$
$$E=mc^2= \mu u \frac{10^{-6} u}{1 \mu u} \frac{931.494 Mev}{1 uc^2}$$

````python
micromassAl26=DATOS["ATOMIC MASS"][155]
def Q(micromassA,micromassb,micromassc):
    Ef=(micromassc+micromassAl26)
    Ei=(micromassA+micromassb)
    MeV=(Ef-Ei)*931.494*1e-06
    return MeV
````

Busquemos ahora reacciones con nucleos $A,b,c$ que cumplan que su nÃºmero atomico $A$ sea:

$$A+b \rightarrow c + {^{26}_{13}Al}$$.

Nota: AquÃ­ se define An

````python
Al=26
nAl=float(13)
zAl=float(13)
#Numero de tamaÃ±o del nucleo considerado An
An=26
A=[]
for a in range(1,An+1):
    for b in range(1,An+1):
        for c in range(1,An+1):
                if a+b-c==Al:
                    A.append([a,b,c])
print("AsÃ­ las posibilidades de [a,b,c] donde 26=a+b-c son nucleos con masa atomica hasta de", An, "son:")
len(A)
````

Ahora, trabajaremos con la primera posibilidad.

````python
print("AsÃ­ para la primera posible reacciÃ³n que produce Al es A+b=c+Al:")
print("A tiene que tener un nÃºmero atÃ³mico de",A[0][0])
print("b tiene que tener un nÃºmero atÃ³mico de",A[0][1])
print("c tiene que tener un nÃºmero atÃ³mico de",A[0][2])
````

Veremos cuales son los nucleos que cumplen la condiciÃ³n para la conservaciÃ³n de protones y neutrones.
Primero filtramos los datos para que satisfagan la combinaciÃ³n anterior.

````python
Condicion_A0=pd.DataFrame(DATOS[DATOS["A"]==float(A[0][0])])
Condicion_A0
````

````python
Condicion_b0=DATOS[DATOS["A"]==float(A[0][1])]
Condicion_b0
````

````python
Condicion_c0=DATOS[DATOS["A"]==float(A[0][2])]
Condicion_c0
````

AÃ±adimos la condiciÃ³n la conservaciÃ³n de n como para z al variar en el dato que buscamos.

````python
for indexA, rowA in Condicion_A0.iterrows():
    for indexb, rowb in Condicion_b0.iterrows():
        for indexc, rowc in Condicion_b0.iterrows():
            if rowA["N"]+rowb["N"]-rowc["N"]==nAl:
                if rowA["Z"]+rowb["Z"]-rowc["Z"]==zAl:
                    print("A index: ",indexA,"C index: ",indexc,"c index: ",indexc)
print("Fin, no hay posibilidades")
````

Ahora variamos en todas las posibilidades que nos brindÃ³ A, determinamos la reacciÃ³n y calculamos Q

````python
INDEX=[]
for k in range(0,len(A)):
    Condicion_A=pd.DataFrame(DATOS[DATOS["A"]==float(A[k][0])])
    Condicion_b=DATOS[DATOS["A"]==float(A[k][1])]
    Condicion_c=DATOS[DATOS["A"]==float(A[k][2])]
# Deterinacion que cumpla con el valance de neutrones y protones
    for indexA, rowA in Condicion_A.iterrows():
        for indexb, rowb in Condicion_b.iterrows():
            for indexc, rowc in Condicion_b.iterrows():
                if rowA["N"]+rowb["N"]-rowc["N"]==nAl:
                    if rowA["Z"]+rowb["Z"]-rowc["Z"]==zAl:
                        INDEX.append([indexA,indexb,indexc])
                        print("ReacciÃ³n:",rowA["A"],rowA["EL"],"+",rowb["A"],rowb["EL"],"=",rowc["A"],rowc["EL"],"+ 26.0 Al")
                        print("Q =: ",Q(rowA["ATOMIC MASS"],rowb["ATOMIC MASS"],rowb["ATOMIC MASS"])," MeV")
                        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
````
