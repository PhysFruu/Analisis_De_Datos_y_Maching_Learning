# Reaccciones Nucleares Binarias 

El proposito de este notebook es el de encontrar todas las reacciones nucleares binarias tal que produzcan ${^{26}_{13}Al}$

$$A+b \rightarrow c + {^{26}_{13}Al}$$.


````python
import pandas as pd
import numpy as np
````

Importamos una base de datos para encontrar la masa y propiedades de los distintos is√≥topos.
Sacada de https://github.com/ffavela/isonav

````python
datos = pd.read_csv("ATOMIC.MASS.csv", sep=',')
print(len(datos))
datos.head()
````

````python
DATOS=datos.drop(labels=0)
#Cambiamos el tipo de dato de esta columna para operar con ella (ya que tiene las masas)
DATOS["ATOMIC MASS"]=DATOS["ATOMIC MASS"].astype("int")
DATOS.head()
````

Buscamos las propiedades del ${^{26}_{}}Al$

````python
DATOS[DATOS["A"]==26]
````

````python
DATOS.loc[155:155]
````

Definimos la funci√≥n para el calculo de energ√≠a.
$$Q=E_f-E_i$$
$$E=mc^2= \mu u \frac{10^{-6} u}{1 \mu u} \frac{931.494 Mev}{1 uc^2}$$

````python
micromassAl26=DATOS["ATOMIC MASS"][155]
def Q(micromassA,micromassb,micromassc):
    Ef=(micromassc+micromassAl26)
    Ei=(micromassA+micromassb)
    MeV=(Ef-Ei)*931.494*1e-06
    return MeV
````

Busquemos ahora reacciones con nucleos $A,b,c$ que cumplan que su n√∫mero atomico $A$ sea:

$$A+b \rightarrow c + {^{26}_{13}Al}$$.

Nota: Aqu√≠ se define $An$ como el tope de la que se busca comparar, en este caso escogemos a 26.

````python
Al=26
nAl=float(13)
zAl=float(13)
#Numero de tama√±o del nucleo considerado An
An=26
A=[]
for a in range(1,An+1):
    for b in range(1,a+1):
        for c in range(1,An+1):
                if a+b-c==Al:
                    A.append([a,b,c])
print("As√≠ las posibilidades de [a,b,c] donde 26=a+b-c son nucleos con masa atomica hasta de", An, "son:")
len(A)
````

Ahora, trabajaremos con la primera posibilidad.

````python
print("As√≠ para la primera posible reacci√≥n que produce Al es A+b=c+Al:")
print("A tiene que tener un n√∫mero at√≥mico de",A[0][0])
print("b tiene que tener un n√∫mero at√≥mico de",A[0][1])
print("c tiene que tener un n√∫mero at√≥mico de",A[0][2])
````

Veremos cuales son los nucleos que cumplen la condici√≥n para la conservaci√≥n de protones y neutrones.
Primero filtramos los datos para que satisfagan la combinaci√≥n anterior.

````python
Condicion_A0=pd.DataFrame(DATOS[DATOS["A"]==float(A[0][0])])
Condicion_A0
````

````python
Condicion_b0=DATOS[DATOS["A"]==float(A[0][1])]
Condicion_b0
````

````python
Condicion_c0=DATOS[DATOS["A"]==float(A[0][2])]
Condicion_c0
````

A√±adimos la condici√≥n la conservaci√≥n de n como para z al variar en el dato que buscamos.

````python
for indexA, rowA in Condicion_A0.iterrows():
    for indexb, rowb in Condicion_b0.iterrows():
        for indexc, rowc in Condicion_c0.iterrows():
            if rowA["N"]+rowb["N"]-rowc["N"]==nAl:
                if rowA["Z"]+rowb["Z"]-rowc["Z"]==zAl:
                    print("A index: ",indexA,"B index: ",indexb,"c index: ",indexc)
````

Ahora variamos en todas las posibilidades que nos brind√≥ A, determinamos la reacci√≥n y calculamos Q

````python
#El indexado se guardo en INDEX de las reacciones posibles
INDEX=[]
for k in range(0,len(A)):
    Condicion_A=pd.DataFrame(DATOS[DATOS["A"]==float(A[k][0])])
    Condicion_b=DATOS[DATOS["A"]==float(A[k][1])]
    Condicion_c=DATOS[DATOS["A"]==float(A[k][2])]
# Deterinacion que cumpla con el balance de neutrones y protones
    for indexA, rowA in Condicion_A.iterrows():
        for indexb, rowb in Condicion_b.iterrows():
            for indexc, rowc in Condicion_c.iterrows():
                if rowA["N"]+rowb["N"]-rowc["N"]==nAl:
                    if rowA["Z"]+rowb["Z"]-rowc["Z"]==zAl:
                        INDEX.append([indexA,indexb,indexc,Q(rowA["ATOMIC MASS"],rowb["ATOMIC MASS"],rowc["ATOMIC MASS"])])
                        print("Reacci√≥n:",rowA["A"],rowA["EL"],"+",rowb["A"],rowb["EL"],"=",rowc["A"],rowc["EL"],"+ 26.0 Al")
                        print("Q =: ",Q(rowA["ATOMIC MASS"],rowb["ATOMIC MASS"],rowc["ATOMIC MASS"])," MeV")
                        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
````

Asi las posibles reacciones binarias de ${^{26}_{13}Al}$ sin contar repeticiones hasta el n√∫cleo con n√∫mero at√≥mico $A_n=26$ son

````python
len(INDEX)
````

Para generar todas las posibles hay que hacer que el factor $A_n$ sea 295, hay que tener cuidado con esto ya que puede tardar bastante correrlo.

````python
#Numero de tama√±o del nucleo considerado An
Anm=295
Am=[]
for a in range(1,Anm+1):
    for b in range(1,a+1):
        for c in range(1,Anm+1):
                if a+b-c==Al:
                    Am.append([a,b,c])
print("As√≠ las posibilidades de [a,b,c] donde 26=a+b-c son nucleos con masa atomica hasta de", Anm, "son:")
len(Am)
````

Ahora habr√≠a que comparar para cada is√≥topo si existe alg√∫no con dicha masa at√≥mica tal que la reacci√≥n sea posible, para esto usando las lineas anteriores, cambiar el $A_n$ por 295, que es la cantidad mayor de masa at√≥mica de los n√∫cleos en la base de datos.

Otra forma de calcularlo para que no se trabe la computadora es realizar el mismo algor√≠tmo pero sin incluir a todos los is√≥topos, fijandonos s√≥lo en los n√∫cleos estables:

````python
#El indexado se guardo en INDEXs de las reacciones posibles
INDEXs=[]
for k in range(0,len(Am)):
    Condicion_As=DATOS[(DATOS["A"]==float(Am[k][0]))&(DATOS["Marcador"]==0)]
    Condicion_bs=DATOS[(DATOS["A"]==float(Am[k][1]))&(DATOS["Marcador"]==0)]
    Condicion_cs=DATOS[(DATOS["A"]==float(Am[k][2]))&(DATOS["Marcador"]==0)]
# Deterinacion que cumpla con el balance de neutrones y protones
    for indexA, rowA in Condicion_As.iterrows():
        for indexb, rowb in Condicion_bs.iterrows():
            for indexc, rowc in Condicion_cs.iterrows():
                if rowA["N"]+rowb["N"]-rowc["N"]==nAl:
                    if rowA["Z"]+rowb["Z"]-rowc["Z"]==zAl:
                        INDEXs.append([indexA,indexb,indexc,Q(rowA["ATOMIC MASS"],rowb["ATOMIC MASS"],rowc["ATOMIC MASS"])])
                        print("Reacci√≥n:",rowA["A"],rowA["EL"],"+",rowb["A"],rowb["EL"],"=",rowc["A"],rowc["EL"],"+ 26.0 Al")
                        print("Q =: ",Q(rowA["ATOMIC MASS"],rowb["ATOMIC MASS"],rowc["ATOMIC MASS"])," MeV")
                        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
````
